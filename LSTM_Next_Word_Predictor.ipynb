```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

- We have to pass our text data as a list in the fit_on_text method of tokenizer since it expects multiple input

```python
tokenizer.word_index # to check which word is encoded which number
```

- Now we will go sentence by sentence and will try to create our dataset for applying supervised learning next word predictor

```python
for sentence in faqs.split('\n'):
  print(tokenizer.texts_to_sequences([sentence])[0]) # This  metohd will assign the index (by word_index) to each word in the sentence
  # Since the senteces were in double list we passed 0 to get the list of senteces encoded
```

- Now we have converted our sentences into numerical data, for equal length of senteces we will do padding

```python
input_sequence=[]
for sentence in faqs.split('\n'):
  tokenized_sequence=tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sequence)):
    input_sequence.append(tokenized_sequence[:i+1])
```

- tokenized_sequence[:i+1] is ngram
- it will take each words and increase the words lengh and for output the next word will be the output
- for the next words we are taking i+1 from the tokenized squence.to create dataset of next word prediction
- Here we have only created a input data

```python
input_sequence
```

- Now the input_sequences are of different lenght so we will perform zero padding. So we will have to find the maximum sentence lenght and according to that we will perform padding

```python
max_length=max([len(x) for x in input_sequence]) # Here we have find the max length of each sentence
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequence,maxlen=max_length,padding='pre') #padding= pre means that we will apply padding on the left side
```

```python
padded_input_sequences
```

- Now for output of next word we will seperate last word from the sentece, since last word we will predict as next sentence form the padded_input_sequence

```python
x=padded_input_sequences[:,:-1] # except the last word
y=padded_input_sequences[:,-1]
```

- Here the output is discrete this means this is not a regression task, it is a classification task
- This is now a milticlass classification

```python
x.shape
```

```python
y.shape
```

```python
# For total words
len(tokenizer.word_index )
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283) # we passed 283 instead of 282, beacause the OneHotEncoding starts encoding from 0 while here in the word_index it starts from 1 so wee added 1 here
```

```python
y # Here since we have taken 283 the fisrt one will always be zero
```

- Now we have our x and y and now we will built our LSTM architecture

- The First layer will be EMbedding layer which will convert the data into Dense Vector representation for every tokens which is better to find similar words in the data

- The Second layer will be of LSTM for the next word predcition

- THe third layer will be of Dense for multiclass classification

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding
```

- Since we have done zero padding our input data becomes sparse data, so to convert from sparse to Dense vector using Embedding
- IN the EMbedding it will convert each word into 100 dimensional dense vectors and total 56 words,  each word of shape (1,56) will become (1,56,100)
- Then each vector will pass in the LSTM which works on timesteps here 56 since there are 56 words in our sentence
- The shape of the LSTM nodes which we will give 150 is the shape of Ht
- Then after the 56th timstep its output is then passed into the Dense layer
- Through that we will get 283 dimensional output in that words with the max occurance will be our final next word predictor answer.

```python
model=Sequential()
model.add(Embedding(283,100,input_length=56)) # the 283 (total vocalbulary count, total no.of words), 100 is the output Dense Vector (The shape of the Embedding), 56 is the output length (since every row contains 56 words),
model.add(LSTM(150)) # In the LSTM we hust have to pass the no.of nodes in the gate it is find accurate by hit and trail
model.add(Dense(283,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=100)
```

- THe training accuracy is 96 which is good
- Now the prediction part we will have to do 3 steps
1. tokenize
2. padd
3. predict

```python
text='mail'
for i in range(5):
  #tokenize
  token_text=tokenizer.texts_to_sequences([text])[0]
  #padding
  padded_token_text=pad_sequences([token_text],maxlen=56,padding='pre')
  # predict
  model.predict(padded_token_text).shape
  pos= np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text=text+" "+word
  print(text)
```
